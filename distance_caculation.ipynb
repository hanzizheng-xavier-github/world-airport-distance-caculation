```python
from geopy.distance import distance
LOCATION_COL = 'Itin'
TARGET_COL = 'Distance'
data[TARGET_COL] = 0.0
for i in range(len(data)):
    total_distance = 0
    if '-' in (data[LOCATION_COL][i]):
        try:
            airport_code_list = data[LOCATION_COL][i].split('-')
            for k in range(len(airport_code_list)):
                sub_total_distance = 0
                airport_code_sublist = airport_code_list[k].split('/')
                for l in range(len(airport_code_sublist)):
                    if l < len(airport_code_sublist) -1:
                        short_distance_dict_list = list()
                        origin_code = airport_code_sublist[l]
                        destination_code = airport_code_sublist[l+1]
                        origin = (
                            float(airport[airport['iata_code'] == origin_code.upper()]
                                  ['latitude_deg']),
                            float(airport[airport['iata_code'] == origin_code.upper()]
                                  ['longitude_deg']),
                        )

                        destination = (
                            float(airport[airport['iata_code'] ==
                                          destination_code.upper()]['latitude_deg']),
                            float(airport[airport['iata_code'] ==
                                          destination_code.upper()]['longitude_deg']),
                        )                 
                        distance_between_2_airports = distance(origin, destination).km
                        sub_total_distance = sub_total_distance + distance_between_2_airports
                total_distance = total_distance + sub_total_distance 
            data[TARGET_COL][i] = total_distance
            print(f'âœ…total distance of trip {i+1} are {total_distance}')
            print('*'*50)
        except:
            pass
                    
    elif '/' in (data[LOCATION_COL][i]):                    
        try:
            airport_code_list = data[LOCATION_COL][i].split('/')
            for j in range(len(airport_code_list)):
                if j < len(airport_code_list) - 1:
                    short_distance_dict_list = list()
                    origin_code = airport_code_list[j]
                    destination_code = airport_code_list[j + 1]
                    origin = (
                        float(airport[airport['iata_code'] == origin_code.upper()]
                              ['latitude_deg']),
                        float(airport[airport['iata_code'] == origin_code.upper()]
                              ['longitude_deg']),
                    )

                    destination = (
                        float(airport[airport['iata_code'] ==
                                      destination_code.upper()]['latitude_deg']),
                        float(airport[airport['iata_code'] ==
                                      destination_code.upper()]['longitude_deg']),
                    )
                    distance_between_2_airports = distance(origin, destination).km
                    info_str = f'the distance between {origin_code} and {destination_code} are {distance_between_2_airports}KM'
                    print(info_str)
                    total_distance = total_distance + distance_between_2_airports
            data[TARGET_COL][i] = total_distance
            print(f'ðŸ‘‰total distance of trip {i+1} are {total_distance}')
            print('*'*50)
        except:
            pass
```
